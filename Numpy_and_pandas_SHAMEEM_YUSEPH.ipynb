PANDAS ACTIVITY

```python
import pandas as pd
import numpy as np

# 1. Load dataset (use seaborn’s Titanic dataset if CSV not available)
import seaborn as sns
titanic = sns.load_dataset("titanic")

# Display first 10 rows
print("\n1. First 10 rows:")
print(titanic.head(10))

# 2. Missing values count
print("\n2. Missing values in each column:")
print(titanic.isnull().sum())

# 3. Percentage survived grouped by gender
print("\n3. Percentage of passengers who survived, grouped by gender:")
survival_gender = titanic.groupby("sex")["survived"].mean() * 100
print(survival_gender)

# 4. Age stats
print("\n4. Age statistics:")
print("Average age:", titanic["age"].mean())
print("Median age:", titanic["age"].median())
print("Std deviation of age:", titanic["age"].std())

# 5. Passenger count by class
print("\n5. Passengers in each class:")
print(titanic["pclass"].value_counts())

# 6. Female, under 18, survived
print("\n6. Female passengers under 18 who survived:")
print(titanic[(titanic["sex"] == "female") & (titanic["age"] < 18) & (titanic["survived"] == 1)])

# 7. Replace missing ages with median
median_age = titanic["age"].median()
titanic["age"].fillna(median_age, inplace=True)
print("\n7. Missing ages replaced with median.")

# 8. Create age_group column
def age_group(age):
    if age < 18:
        return "Child"
    elif age <= 60:
        return "Adult"
    else:
        return "Senior"

titanic["age_group"] = titanic["age"].apply(age_group)
print("\n8. Age groups created:")
print(titanic[["age", "age_group"]].head(10))

# 9. Top 5 by highest fare
print("\n9. Top 5 passengers by highest fare:")
print(titanic.sort_values("fare", ascending=False).head(5))

# 10. Average fare grouped by class & gender
print("\n10. Average fare grouped by pclass and sex:")
print(titanic.groupby(["pclass", "sex"])["fare"].mean())

# 11. Rename columns
titanic.rename(columns={"sex": "gender", "fare": "ticket_fare"}, inplace=True)
print("\n11. Columns renamed. New column names:")
print(titanic.columns)

# 12. Survival count by class
print("\n12. Survival count grouped by class:")
print(titanic.groupby("pclass")["survived"].value_counts())

# 13. Average age grouped by class and gender
print("\n13. Average age by class and gender:")
print(titanic.groupby(["pclass", "gender"])["age"].mean())
```

```python
import numpy as np

# ----------------------------------------------------------------
# 1. Array from 10 to 49, reshape into 4x10, extract last column
arr = np.arange(10, 50)       # values from 10 to 49
matrix = arr.reshape(4, 10)   # reshape 4x10
last_col = matrix[:, -1]      # extract last column
print("1. Reshaped 4x10 matrix:\n", matrix)
print("Last column:\n", last_col)

# ----------------------------------------------------------------
# 2. Two 3x3 matrices (all 5s and 1–9)
A = np.full((3, 3), 5)                # all 5s
B = np.arange(1, 10).reshape(3, 3)    # values 1–9
add = A + B
multiply = A * B
transpose = multiply.T
print("\n2. Matrix A:\n", A)
print("Matrix B:\n", B)
print("A + B:\n", add)
print("A * B (element-wise):\n", multiply)
print("Transpose of result:\n", transpose)

# ----------------------------------------------------------------
# 3. Random 15 integers, replace >50 with -1, count replaced
rand_arr = np.random.randint(10, 101, size=15)
replaced = np.where(rand_arr > 50, -1, rand_arr)
count_replaced = np.sum(rand_arr > 50)
print("\n3. Original array:\n", rand_arr)
print("After replacement:\n", replaced)
print("Count replaced:", count_replaced)

# ----------------------------------------------------------------
# 4. 5x4 random array (0–1), row with max sum, normalize rows
arr = np.random.rand(5, 4)    # random values 0–1
row_sums = arr.sum(axis=1)
max_row_index = np.argmax(row_sums)
normalized = arr / arr.sum(axis=1, keepdims=True)
print("\n4. Random 5x4 array:\n", arr)
print("Row sums:", row_sums)
print("Row with max sum (index):", max_row_index)
print("Normalized array:\n", normalized)

# ----------------------------------------------------------------
# 5. Find unique elements and duplicates
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique_vals = np.unique(arr)
duplicates = np.array([x for x in unique_vals if np.sum(arr == x) > 1])
print("\n5. Unique elements:", unique_vals)
print("Duplicates:", duplicates)

# ----------------------------------------------------------------
# 6. 6x6 array with 1–36, extract middle 4x4, then even numbers
matrix = np.arange(1, 37).reshape(6, 6)
submatrix = matrix[1:5, 1:5]          # middle 4x4
even_numbers = submatrix[submatrix % 2 == 0]
print("\n6. 6x6 matrix:\n", matrix)
print("Middle 4x4 submatrix:\n", submatrix)
print("Even numbers from submatrix:", even_numbers)

import numpy as np

# ----------------------------------------------------------------
# 7. Generate (3,4), reshape to (2,6), flatten, reshape (4,3)
arr = np.arange(12).reshape(3, 4)
reshaped = arr.reshape(2, 6)
flattened = reshaped.flatten()
back_to_4x3 = flattened.reshape(4, 3)
print("7. Original (3x4):\n", arr)
print("Reshaped (2x6):\n", reshaped)
print("Flattened 1D:", flattened)
print("Back to (4x3):\n", back_to_4x3)

# ----------------------------------------------------------------
# 8. Append new row and new column
a = np.array([[1, 2], [3, 4]])
a = np.vstack([a, [5, 6]])             # append new row
a = np.column_stack([a, [7, 8, 9]])    # append new column
print("\n8. After appending row & column:\n", a)

# ----------------------------------------------------------------
# 9. Join arrays, reshape 3x3, extract diagonal
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
joined = np.concatenate([a, b, c])
matrix = joined.reshape(3, 3)
diagonal = np.diag(matrix)
print("\n9. Joined array:", joined)
print("Reshaped 3x3 matrix:\n", matrix)
print("Diagonal elements:", diagonal)

# ----------------------------------------------------------------
# ----------------------------------------------------------------
# 10. Numbers 1–20 divisible by 3 or 5
arr = np.arange(1, 21)
divisible = arr[(arr % 3 == 0) | (arr % 5 == 0)]
count = divisible.size
print("\n10. Numbers divisible by 3 or 5:", divisible)
print("Count:", count)
```
