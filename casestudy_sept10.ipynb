```python
import pandas as pd
pop = pd.read_csv("/content/population.csv")
pop
```

```python
import pandas as pd
gdp = pd.read_excel("/content/gdp.xlsx")
gdp

```

```python
import pandas as pd
ius = pd.read_json("/content/internet_users.json")
ius
```

```python
import pandas as pd
lit = pd.read_xml("/content/literacy_rate.xml")
lit
```

```python
pop.rename(columns={'country':'name'}, inplace=True)
pop
```

```python
gdp.rename(columns={'Country':'name'}, inplace=True)
gdp
```

```python
ius.rename(columns={'country':'name'}, inplace=True)
ius
```

```python
pop_gdp = pd.merge(pop,gdp, on='name', how='inner')
pop_gdp
```

```python
pop_gdp_ius = pd.merge(pop_gdp,ius, on='name', how='inner')
pop_gdp_ius
```

```python
pop_gdp_ius_lit = pd.merge(pop_gdp_ius,lit, on='name', how='inner')
pop_gdp_ius_lit
```

```python
import pandas as pd


pop_gdp_ius_lit["internet_penetration_rate"] = (pop_gdp_ius_lit["internet_users"] / pop_gdp_ius_lit["population"] * 100)

pop_gdp_ius_lit
```

```python
high = pop_gdp_ius_lit.loc[pop_gdp_ius_lit['internet_penetration_rate'].idxmax()]
high
```

```python
avg = pop_gdp_ius_lit['literacy_rate'].mean()
print(avg)
```

```python
pop_gdp_ius_lit[['population','literacy_rate','internet_users']].corr()
```

```python
import sqlite3

conn = sqlite3.connect("merge")
cursor = conn.cursor()
pop_gdp_ius_lit.to_sql(name='merge', con=conn, if_exists='replace', index=False)
```

```python
pd.read_sql_query('select * from merge',conn)
```

```python
pd.read_sql_query('select name , internet_penetration_rate from merge order by internet_penetration_rate desc limit 1',conn)
```

```python
pd.read_sql_query('select avg(literacy_rate) from merge',conn)
```

```python
pop_gdp_ius_lit[['literacy_rate','internet_penetration_rate']].corr()
```

```python
pd.read_sql_query('select name,GDP/population AS per from merge',conn)
```

```python
pd.read_sql_query('select name,per from (select name ,GDP/population as per from merge) where per > 10000',conn)
```

```python
pd.read_sql_query('select sum(population) from merge',conn)
```

```python
pd.read_sql_query('select name , literacy_rate,internet_penetration_rate from merge order by literacy_rate asc limit 1',conn)
```
